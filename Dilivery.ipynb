```python
import random
from datetime import datetime, timedelta
import pandas as pd
from faker import Faker


fake = Faker()

# Iraq-Specific Matrices 
iraqi_cities = [
    "Baghdad",
    "Basra",
    "Karbala",
    "Erbil",
    "Najaf",
    "Diala",
    "Salah_Aldeen",
]
Vehicles = ["Car", "Motorcycle"]
Status_Options = ["Cancelled by Customer", "Cancelled by Captin", "Compeleted"]
Payment_Options = ["Cash", "Qi Card", "Zain Cash", "MasterCard"]

# 1. Generate a Dataset of 5,000 Customers
Customers = []
for i in range(1, 5001):
    Customers.append(
        {
            "User_id": i,
            "User_name": fake.name(),
            "phone_number": f"+9647{random.randint(5, 9)}{random.randint(1000000, 9999999)}",
            "City": random.choice(iraqi_cities),
            "Join_Date": fake.date_between(start_date="-2y", end_date="today"),
        }
    )
df_Customers = pd.DataFrame(Customers)

# 2. Generate 300 Delivery Drivers
drivers = []
for i in range(1, 301):
    drivers.append(
        {
            "driver_id": i,
            "driver_name": fake.name(),
            "phone_number": f"+9647{random.randint(5, 9)}{random.randint(1000000, 9999999)}",
            "vehicle_type": random.choice(Vehicles),
            "rating": round(random.uniform(3.5, 5.0), 1),
            "registration_date": fake.date_between(
                start_date="-2y", end_date="today"
            ),
        }
    )
df_drivers = pd.DataFrame(drivers)

# 3. Generate 15,000 Simulated Trips
trips = []
start_date = datetime(2025, 1, 1)

for i in range(1, 15001):
    random_days = random.randint(0, 500)
    trip_time = start_date + timedelta(
        days=random_days,
        hours=random.randint(0, 23),
        minutes=random.randint(0, 59),
    )

    # اختيار حالة الرحلة بناءً على الاحتمالات
    status = random.choices(Status_Options, weights=[10, 5, 85])[
        0
    ]  

    # حساب السعر بناءً على حالة الرحلة
    fare = random.randint(3, 10) * 1000 if status == "Compeleted" else 0

    trips.append(
        {
            "trip_id": i,
            "User_id": random.randint(1, 5000),
            "driver_id": random.randint(1, 300),
            "trip_date": trip_time,
            "trip_status": status,
            "fare_iqd": fare,
            "payment_method": random.choice(Payment_Options)
            if status == "Compeleted"
            else "N/A",
        }
    )

df_trips = pd.DataFrame(trips)

# 4. Export Data to CSV
df_drivers.to_csv("drivers.csv", index=False, encoding="utf-8-sig")
df_Customers.to_csv("customers.csv", index=False, encoding="utf-8-sig")  
df_trips.to_csv("trips.csv", index=False, encoding="utf-8-sig")

print("Done! All 3 CSV files have been generated successfully.")
```
